```python
import numpy as np
import netCDF4
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ncfile = netCDF4.Dataset('../../miercoles/era.slp.2007.nc','r')  # Open file

T = ncfile.variables["msl"]   # Get Variable object for "temperature"

print(type(T))

T_data = T[:]          # Extract NumPy data array from Variable

print(type(T_data))
print(T.ncattrs())      # List all variable attributes

T.units                # Units attribute
```

```python
#ncf = netCDF4.Dataset("..............", "r")
lon = ncfile.variables["longitude"][:]
lat = ncfile.variables["latitude"][:]
temp_data = ncfile.variables["msl"][0,:,:]
temp_data = temp_data/100
temp_data.shape
```

```python
## GRAFICAMOS LA PRESION AL NIVEL DEL MAR EN PASCALES
#####################################################
from mpl_toolkits.basemap import Basemap
# llcrnrlat,llcrnrlon,urcrnrlat,urcrnrlon
# are the lat/lon values of the lower left and upper right corners
# of the map.
# resolution = 'c' means use crude resolution coastlines.
m = Basemap(projection='cyl',llcrnrlat=-60,urcrnrlat=-7.5,\
            llcrnrlon=259.5,urcrnrlon=352,resolution='c')
m.drawcoastlines()
m.drawcountries()
# draw parallels and meridians.
m.drawparallels(np.array([-60, -50, -40, -30, -20, -10]), labels=[1,0,0,1])
m.drawmeridians(np.array([260, 280, 300, 320, 340]), labels=[1,0,0,1])
plt.title("Presion al nivel del mar (Pa)")

# ACA DIBUJO LOS CONTORNOS
plt.contourf(lon, lat, temp_data, 10, cmap=plt.cm.rainbow)
plt.axis([259.5, 352.5, -60, -7.5])
plt.colorbar(orientation="horizontal")

cont = plt.contour(lon, lat, temp_data, 10)   # Save object
plt.clabel(cont, fmt='%.1f')
plt.show()
plt.savefig('test.png')
```

```python
#PROYECCION CON BASEMAP

from mpl_toolkits.basemap import Basemap
# llcrnrlat,llcrnrlon,urcrnrlat,urcrnrlon
# are the lat/lon values of the lower left and upper right corners
# of the map.
# resolution = 'c' means use crude resolution coastlines.
m = Basemap(projection='cyl',llcrnrlat=-60,urcrnrlat=-7.5,\
            llcrnrlon=259.5,urcrnrlon=352,resolution='c')
m.drawcoastlines()
m.drawcountries()
##m.fillcontinents(color='coral',lake_color='aqua')
# draw parallels and meridians.
m.drawparallels(np.arange(-60.,-7.5,10), labels=[1,0,0,1])
m.drawmeridians(np.arange(259.5,352.,10), labels=[0,1,0,1])
##m.drawmapboundary(fill_color='aqua')
plt.title("Equidistant Cylindrical Projection")
plt.show()
```
